```python
from ultralytics import YOLO
modelM_50 = YOLO("yolo11m.pt")

modelM_50.train(
    data="/content/dataset/data.yaml",
    epochs=50,
    imgsz=512,
    batch=4,
    device=0,
    name="m_epochs_50",
    workers=2
)
```

```python
model = YOLO("/content/runs/detect/m_epochs_502/weights/best.pt")
model.predict(
    source="/content/dataset/test/images",
    imgsz=512,
    conf=0.25,
    save=True,
    project="/content/runs/detect",
    name="m50_infer"
)
```
